```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time

# Import API key
from api_keys import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "weather_data.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)
lat_lngs
```

```python
# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
#Create base url for API calls
base_url = "http://api.openweathermap.org/data/2.5/weather?"
#create params for call
params = {
    "q":"",
    "units":"imperial",
    "APPID":api_key
}
```

```python
# function which gets weather data for city and prints progress. Returns "ERROR" when unable to retrieve.
def get_weather(city):
    params["q"] = city
    try:
        print("Finding weather for the city of {}".format(city))
        response = requests.get(url = base_url,params = params).json()
        return response
    except:
        return "ERROR"
```

```python
#results list to hold json response given by function get_weather
results = []
print("-------------------\nBeginning Data Retrieval\n-------------------")
#Loop through cities to append weather data to results and check for any errors to skip them and print to screen
for city in cities:
    check = get_weather(city)
    if (check == "ERROR") or (check.get("cod") != 200):
        print("Unable to retrieve data. Skipping.")
        continue
    else:
        results.append(check)
    
print("-------------------\nData Retrieval Complete\n-------------------")
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
#fill dictionary with data from results
weather_dict = {
    "City":[result.get("name") for result in results],
    "Cloudiness":[result.get("clouds").get("all") for result in results],
    "Country":[result.get("sys")["country"] for result in results],
    "Date":[result.get("dt") for result in results],
    "Humidity":[result.get("main").get("humidity") for result in results],
    "Lat":[result.get("coord").get("lat") for result in results],
    "Lng":[result.get("coord").get("lon") for result in results],
    "Max_Temp":[result.get("main").get("temp_max") for result in results],
    "Wind_Speed":[result.get("wind").get("speed") for result in results]
}
```

```python
# place weather_dict into df and output to csv
weather_df = pd.DataFrame(weather_dict)
weather_df.to_csv(output_data_file)
```

### Plotting the Data
* Use proper labeling of the plots using plot titles (including date of analysis) and axes labels.
* Save the plotted figures as .pngs.

#### Latitude vs. Temperature Plot

```python
plt.scatter(x=weather_df["Lat"],y=weather_df["Max_Temp"])
plt.grid()
plt.xlabel("Latitude")
plt.ylabel("Max Temperature (F)")
plt.title("City Latitude vs Max Temperature for "+time.strftime("%d %b %Y", time.gmtime()))
plt.savefig("LatVsTemp.png")
plt.show()
```

```python
plt.scatter(x=weather_df["Lat"],y=weather_df["Humidity"])
plt.grid()
plt.xlabel("Latitude")
plt.ylabel("Humidity (%)")
plt.title("City Latitude vs Humidity for"+time.strftime("%d %b %Y", time.gmtime()))
plt.savefig("LatVsHumidity.png")
plt.show()
```

#### Latitude vs. Cloudiness Plot

```python
plt.scatter(x=weather_df["Lat"],y=weather_df["Cloudiness"])
plt.grid()
plt.xlabel("Latitude")
plt.ylabel("Cloudiness (%)")
plt.title("City Latitude vs Cloudiness for"+time.strftime("%d %b %Y", time.gmtime()))
plt.savefig("LatVsCloudiness.png")
plt.show()
```

#### Latitude vs. Wind Speed Plot

```python
plt.scatter(x=weather_df["Lat"],y=weather_df["Wind_Speed"])
plt.grid()
plt.xlabel("Latitude")
plt.ylabel("Wind Speed (mph)")
plt.title("City Latitude vs Wind Speed for"+time.strftime("%d %b %Y", time.gmtime()))
plt.savefig("LatVsWind.png")
plt.show()
```
